```python
#Analysis of Amazon stock only
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("AMZN.csv")
Price=np.array(df['Open'])
Time=np.array(df['Date'])
Volume=np.array(df['Volume'])

plt.subplot(2,1,1)
plt.title('Stock Price v/s Time')
plt.plot(Time,Price)
plt.xlabel('Time')
plt.ylabel('Open Stock Price')
plt.show()

plt.subplot(2,1,2)
plt.title('Volume Traded v/s Time')
plt.plot(Time,Volume)
plt.xlabel('Time')
plt.ylabel('Volume Traded')
plt.show()

moving=df['Open'].rolling(50).mean()#50 day window
plt.title('Moving Average v/s Time')
plt.plot(Time,moving)
plt.xlabel('Time')
plt.ylabel('Moving Average')
plt.show()

df['Trend']=df['Close']-df['Open']
daily_avg=df['Trend'].mean()
print('The daily return average of a stock is:',daily_avg)
tlabel=list(set(df.Trend))
freq={}
for i in tlabel:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1
labels = []
sizes = []
for x, y in freq.items():
    labels.append(x)
    sizes.append(y)
plt.pie(sizes, labels=labels,rotatelabels=True)
plt.axis('equal')
plt.legend()#Since uniqueness in the trend column is very high,so pie chart is not a good option since huge number of labels are involved and overcrowding happens.
plt.show()
```

```python
#Combined Analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("AMZN.csv")
df2=pd.read_csv("AAPL.csv")
df3=pd.read_csv("MSFT.csv")
df4=pd.read_csv("GOOGL.csv")

Price=np.array(df['Open'])
Price2=np.array(df2['Open'])
Price3=np.array(df3['Open'])
Price4=np.array(df4['Open'])

Time=np.array(df['Date'])

Volume=np.array(df['Volume'])
Volume2=np.array(df2['Volume'])
Volume3=np.array(df3['Volume'])
Volume4=np.array(df4['Volume'])

moving=df['Open'].rolling(50).mean()#50 day window
moving2=df2['Open'].rolling(50).mean()#50 day window
moving3=df3['Open'].rolling(50).mean()#50 day window
moving4=df4['Open'].rolling(50).mean()#50 day window

#daily returns of stock
df['Trend']=df['Close']-df['Open']
df2['Trend']=df2['Close']-df2['Open']
df3['Trend']=df3['Close']-df3['Open']
df4['Trend']=df4['Close']-df4['Open']
daily_avg=df['Trend'].mean()
daily_avg2=df2['Trend'].mean()
daily_avg3=df3['Trend'].mean()
daily_avg4=df4['Trend'].mean()
print('The daily return average of Amazon stock is:',daily_avg)
print('The daily return average of Apple stock is:',daily_avg2)
print('The daily return average of Microsoft stock is:',daily_avg3)
print('The daily return average of Google stock is:',daily_avg4)

plt.subplot(2,2,1)
plt.title('Stock Price v/s Time')
plt.plot(Time,Price,color='Green',label='AMZN')
plt.plot(Time,Price2,color='Red',label='AAPL')
plt.plot(Time,Price3,color='Blue',label='MSFT')
plt.plot(Time,Price4,color='Pink',label='GOOGL')
plt.xlabel('Time')
plt.ylabel('Open Stock Price')
plt.legend(loc='upper left')
plt.show()

plt.subplot(2,2,2)
plt.title('Volume Traded v/s Time')
plt.plot(Time,Volume,color='Green',label='AMZN')
plt.plot(Time,Volume2,color='Red',label='AAPL')
plt.plot(Time,Volume3,color='Blue',label='MSFT')
plt.plot(Time,Volume4,color='Pink',label='GOOGL')
plt.xlabel('Time')
plt.ylabel('Volume Traded')
plt.legend(loc='upper left')
plt.show()

plt.subplot(2,2,3)
plt.title('Moving Average v/s Time')
plt.plot(Time,moving,label='AMZN')
plt.plot(Time,moving2,label='AAPL')
plt.plot(Time,moving3,label='MSFT')
plt.plot(Time,moving4,label='GOOGL')
plt.xlabel('Time')
plt.ylabel('Moving Average')
plt.legend(loc='upper left')
plt.show()

plt.subplot(2,2,4)
plt.title('Daily Returns of Stocks v/s Time')
plt.plot(Time,df['Trend'],label='AMZN')
plt.plot(Time,df2['Trend'],label='AAPL')
plt.plot(Time,df3['Trend'],label='MSFT')
plt.plot(Time,df4['Trend'],label='GOOGL')
plt.xlabel('Time')
plt.ylabel('Daily Returns')
plt.legend(loc='upper left')
plt.show()
```
